##Matplotlib

```python
tiempo = np.linspace(0, 60, 180) # Creo una lista de 180 puntos de tiempo entre 0 y 60 segundos
crecimiento1 = 5*tiempo-10*np.random.rand(180) # Calculo el primer crecimiento con un poco de ruido al azar
crecimiento2 = 5.8*tiempo-12*np.random.rand(180) # Calculo el crecimiento del segundo árbol (crece un poco más rápido)
```

```python
!pip install matplotlib # Instalo la librería matplotlib para hacer gráficas
import matplotlib.pyplot as plt # La importo con el nombre 'plt'
```

```python
plt.plot(tiempo, crecimiento1)# Dibujo la línea para el primer árbol
plt.plot(tiempo, crecimiento2 ,color = "red") # Dibujo la línea para el segundo árbol y la pongo de color rojo

# Le pongo nombre al eje horizontal (X) y al eje vertical (Y)
plt.xlabel("Tiempo (s)")
plt.ylabel("Crecimiento (cm)")

# Le pongo un título
plt.title("Grafico de Crecimiento")
plt.grid() # Le pongo una cuadrícula al fondo
plt.legend(("Arbol 1","Arbol 2")) # Pongo una leyenda para saber cuál línea es cuál árbol
plt.show() # mostrar la gráfica
```

##Pandas

```python
import pandas as pd # Importo la librería Pandas con el nombre 'pd'
from pandas import Series, DataFrame # Traigo herramientas específicas para crear listas (Series) y tablas (DataFrame)
```

```python
# Creo dos filas con 4 columnas
obj = pd.Series([[4, 7, -5, 3],[6, 8 , 5 , 2]])
print (obj)
```

```python
print(obj[0]) # Cojo la primero fila (0)
```

```python
# Creo una serie dándole nombres a cada fila (a0, a1, a2, a3)
obj = pd.Series((4, 7, -5, 3),("a0","a1","a2", "a3"))
print (obj)
```

```python
print(obj[1]) # Consulto el valor que está en la fila 2 (a1)
```

```python
obj["a4"] = -9 # Añado un nuevo dato con la etiqueta "a4" y le asigno el valor -9
```

```python
print(obj) # Muestro cómo quedó la serie con el nuevo valor agregado
```

##DataFrame

```python
# Armo un diccionario con la información de estados, años y población
data = {"state": ["Ohio", "Ohio", "Ohio", "Nevada", "Nevada", "Nevada"],
"year": [2000, 2001, 2002, 2001, 2002, 2003],
"pop": [1.5, 1.7, 3.6, 2.4, 2.9, 3.2]}
```

```python
df = pd.DataFrame(data)# Convierto ese diccionario en una tabla formal (DataFrame) llamada 'df'
```

```python
df.head() # Muestro las primeras 5 filas de la tabla para ver cómo quedó
```

```python
df["state"]# Extraigo solamente la columna de los nombres de los estados
```

```python
df["year"][0]# Saco el primer año de la columna "year" (posición 0)
```

```python
df["state"][2]# Veo qué estado hay en la tercera fila (posición 2) de la columna "state"
```
