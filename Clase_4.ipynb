```python
import matplotlib.pyplot as plt # Importa la librería matplotlib.pyplot y la abrevia como plt para crear gráficos y visualizaciones
```

```python
plt.figure(figsize=(10,10)) # Aquí creo el espacio donde voy a hacer el gráfico y le doy un tamaño de 10x10

# Aquí hago un histograma con la columna 'CREDIT'
# uso 50 barras para dividir los datos
# le pongo color violeta
# ajusto el ancho de las barras
# y les pongo borde negro para que se distingan mejor
plt.hist(data['CREDIT'], bins=50, color='#EE82EE', rwidth=0.7, edgecolor='black')

# Aquí calculo el promedio de los valores de 'CREDIT' y lo guardo en la variable promedio
promedio = data['CREDIT'].mean()

# Aquí dibujo una línea vertical donde está el promedio
# la hago roja, punteada y un poco gruesa
# también le pongo una etiqueta mostrando el valor del promedio con 2 decimales
plt.axvline(promedio, color='red', linestyle='dashed', linewidth=2, label=f'Media: {promedio:.2f}')
plt.xticks(rotation=40)
# Roto las etiquetas del eje X 40 grados para que se vean mejor
```

```python
import seaborn as sns # Aquí importo la librería seaborn y la nombro como sns para poder hacer gráficos más bonitos y fáciles de personalizar
```

```python
plt.figure(figsize=(10,10)) # Aquí creo el espacio del gráfico y le doy un tamaño grande para que se vea mejor

sns.histplot(data['CREDIT'], bins= 20,)# Aquí hago un histograma básico de la columna 'CREDIT' con 20 barras
sns.set_theme(style="whitegrid")# Aquí aplico un estilo de fondo con cuadrícula blanca

# 2. Usar histplot con los parámetros de tu imagen
sns.histplot(data=data, x='CREDIT', bins=50,
             color='orchid',      # Color similar al violeta de tu foto
             edgecolor='black',   # Aquí pongo bordes negros a las barras
             shrink=0.8,          # Esto separa las barras (de tu documentación)
             kde=True)            # Añade la curva de densidad suave

# 3. La línea de la media (manteniendo lo que ya hiciste)
plt.axvline(data['CREDIT'].mean(), color='red', linestyle='--', label='Media')# Aquí dibujo una línea roja punteada (promedio)

plt.title('Distribución Profesional de Créditos', fontsize=15)# Aquí pongo el título del gráfico y aumento el tamaño de la letra

plt.xticks(rotation=40)# Aquí giro los valores del eje X para que se vean más ordenados
plt.legend()# Aquí muestro la leyenda (donde aparece la línea de la media)
plt.show()# Aquí finalmente muestro el gráfico en pantalla
```

```python
import seaborn as sns #Importo seaborn
import matplotlib.pyplot as plt #Importo matplotlib

# 1. Configuramos el estilo de fondo
sns.set_theme(style="ticks")

# 2. Creamos el lienzo (tamaño de la imagen)
plt.figure(figsize=(12, 6))

# 3. Dibujamos el Boxplot con el truco para la línea central
sns.boxplot(
    data=data,
    x='FAM_MEMBERS',
    y='CREDIT',
    hue='FAM_MEMBERS',
    palette="pastel",          # Usamos colores pastel para que la línea resalte más
    legend=False,              # Aquí quito la leyenda porque no es necesaria
    # 'medianprops' cambia la línea de adentro: color rojo y más gruesa
    medianprops={"color": "red", "linewidth": 2}
)

# 4. Limpiamos los ejes para que se vea profesional (como en tu imagen)
sns.despine(offset=10, trim=True)

# 5. Formato extra para que sea fácil de leer
plt.title("Distribución de Crédito por Miembros de la Familia", fontsize=14, fontweight='bold') # Aquí pongo el título del gráfico en negrita
plt.ylabel("Monto del Crédito")# Aquí nombro el eje Y
plt.xlabel("Miembros de la Familia")# Aquí nombro el eje X

# Esto quita la notación científica (el 1e6) para ver los números completos
plt.ticklabel_format(style='plain', axis='y')
# Aquí hago que los números del eje Y se vean completos y no en notación científica

plt.show() # Aquí muestro el gráfico final
```

**1. La parte de abajo de la caja**

El 25% de las familias (las que pidieron poquito) se quedan debajo de esta línea.

**2. La línea de en medio**

Es la mediana. Nos dice que exactamente la mitad de las familias están por debajo de ella y la otra mitad está por arriba.

**3. La parte de arriba de la caja**

Por encima de esta línea solo queda el 25% de las familias que pidieron más dinero (sin contar a los "exagerados" de los circulitos).

**4. El cuerpo de la caja (Lo que tiene color)**

Todo el espacio que hay dentro de la caja (entre la puerta de entrada y la de salida) es donde vive el 50% de la gente. Es decir, la mitad de todas las familias están apretaditas ahí adentro.

Si la caja es muy alta: Significa que las familias de ese grupo son muy diferentes entre sí.

Si la caja es muy bajita: Significa que casi todas las familias piden casi la misma cantidad de dinero.

**5. Los "bigotes" (Las líneas que salen hacia arriba y abajo)**

Son como los brazos de la caja. Llegan hasta donde están las familias "normales" que pidieron lo más poquito y lo más mucho.
